#  Derive a new column called agegroup from age column where persons falling in different age ranges are categorized as below

```python
df["age"].min()
df
```

```python
df["age"].max()
```

```python
def age_colum(age):
    if age<=15 :
        return "child"
    elif 15<age<=35 :
        return "young"
    elif 35<age<55 :
        return "middle age "
    else :
        return "old"
        
        
```

```python
df["agegroup"]=df["age"].apply(age_colum)
df["agegroup"].unique()
```

```python
df
```
