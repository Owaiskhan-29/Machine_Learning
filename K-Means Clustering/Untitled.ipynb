```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline

# Step 1: Read the data
data_df = pd.read_csv("Cust_Spend_Data_New.csv") 

# Step 2: Preprocess the data
cust_df = data_df.drop(['Name', 'Cust_ID'], axis=1)

# Step 3: Standardize the features using StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(cust_df)

# Step 4: Perform K-means clustering with multiple initializations
k_means = KMeans(n_clusters=3, n_init=5)
k_means.fit(scaled_df)

# Step 5: Get the cluster labels for each data point
cluster_labels = k_means.labels_

# Step 6: Add cluster labels to the original data
data_df["Clus_kmeans"] = cluster_labels

# Step 7: Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(x=scaled_df[:, 0], y=scaled_df[:, 1], c=cluster_labels, cmap='viridis')
plt.title('K-means Clustering')
plt.xlabel('Scaled Avg_Mthly_Spend')
plt.ylabel('Scaled No_Of_Visits')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline

# Step 1: Load the Iris dataset
iris = load_iris()

# Convert the dataset into a pandas DataFrame
data_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data_df['target'] = iris.target

# Step 2: Preprocess the data
X = data_df.drop('target', axis=1)

# Step 3: Standardize the features using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

# Step 4: Perform K-means clustering
k_means = KMeans(n_clusters=3, random_state=42)
k_means.fit(scaled_X)

# Step 5: Get the cluster labels for each data point
cluster_labels = k_means.labels_

# Step 6: Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue=cluster_labels, data=data_df)
plt.title('K-means Clustering')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

# Step 1: Read the dataset
data_df = pd.read_csv("Mall_Customers.csv")

# Display the first few rows of the data
data_df.head()

# Step 2: Preprocess the data
X = data_df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Step 3: Visualize the data before K-means clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data_df)
plt.title('Data Before K-means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Step 4: Standardize the features using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

# Step 5: Visualize the scaled data
scaled_df = pd.DataFrame(data=scaled_X, columns=['Annual Income (k$)', 'Spending Score (1-100)'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=scaled_df)
plt.title('Scaled Data Before K-means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Step 6: Perform K-means clustering
k_means = KMeans(n_clusters=5, random_state=42)
k_means.fit(scaled_X)

# Step 7: Get the cluster labels for each data point
cluster_labels = k_means.labels_

# Step 8: Visualize the K-means clustering results
data_df['Cluster'] = cluster_labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data_df, palette='viridis')
plt.title('K-means Clustering Results')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
