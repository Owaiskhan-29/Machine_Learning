```python
#Import necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Step 1: Read the dataset
data_df = pd.read_csv("Mall_Customers.csv")
```

```python
# Display the first few rows of the data
data_df
```

```python
#Summary of the data
```

```python
data_df.shape
```

```python
# Get descriptive statistics of the data
data_df.describe()
```

```python
# Step 2: Preprocess the data

# Select the relevant columns
X = data_df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Step 3: Visualize the data before K-means clustering

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data_df)
plt.title('Data Before K-means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Step 4: Standardize the features using StandardScaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
# Step 6: Perform K-means clustering
k_means = KMeans(n_clusters=5, random_state=42)
k_means.fit(scaled_X)
```

```python
# Step 7: Get the cluster labels for each data point
cluster_labels = k_means.labels_
```

```python
# Step 8: Visualize the K-means clustering results
data_df['Cluster'] = cluster_labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data_df, palette='viridis')
plt.title('K-means Clustering Results')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
wss=[]
```

```python
for k in range(2,11):
    KM = KMeans(n_clusters=k)
    KM.fit(scaled_df)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
# Plot the WSS values
plt.xlabel("K")
plt.ylabel('Sum of squared error')
plt.plot(range(2,11), wss)
```
